```python
df_clean = df.copy()
```

```python
df_clean = df_clean.dropna(subset=['sellingprice', 'saledate'])
```

```python
df_clean = df_clean.drop(columns=['vin'])
```

```python
# Fill numerical columns properly
df_clean['condition'] = df_clean['condition'].fillna(df_clean['condition'].median())

df_clean['odometer'] = df_clean['odometer'].fillna(df_clean['odometer'].median())

df_clean['mmr'] = df_clean['mmr'].fillna(df_clean['mmr'].median())
```

```python
cols = ['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior']

df_clean[cols] = df_clean[cols].fillna('Unknown')
```

```python
df_clean['saledate'] = pd.to_datetime(df_clean['saledate'], errors='coerce', utc=True)
```

```python
df_clean['saledate'].head()
```

```python
df_clean['saledate'].isnull().sum()
```

```python
df_clean = df_clean.dropna(subset=['saledate'])
```

```python
df_clean.isnull().sum()
```

```python
df_clean.shape
```

```python
df_clean.to_csv("../data/processed/clean_data.csv", index=False)
```

# 🔵 Advanced Analysis (Level 2)

```python
# Create new feature: car age
df ['car_age'] = 2025 - df['year']

# Check result
df[['year', 'car_age']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.scatterplot(x='car_age', y='sellingprice', data=df)

plt.title("Car Age vs Selling Price")
plt.xlabel("Car Age")
plt.ylabel("Selling Price")

plt.savefig("../outputs/charts/car_age_vs_price.png")
plt.show()
```

## Removing Outliers

```python
# Remove extreme values (top 1%)
df = df[df['sellingprice'] < df['sellingprice'].quantile(0.99)]

df.shape
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['sellingprice'], bins=50)

plt.title("Price Distribution (After Outlier Removal)")

plt.savefig("../outputs/charts/price_distribution_cleaned.png")
plt.show()
```
