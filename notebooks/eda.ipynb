```python
df_clean = df.copy()
```

```python
df_clean = df_clean.dropna(subset=['sellingprice', 'saledate'])
```

```python
df_clean = df_clean.drop(columns=['vin'])
```

```python
# Fill numerical columns properly
df_clean['condition'] = df_clean['condition'].fillna(df_clean['condition'].median())

df_clean['odometer'] = df_clean['odometer'].fillna(df_clean['odometer'].median())

df_clean['mmr'] = df_clean['mmr'].fillna(df_clean['mmr'].median())
```

```python
cols = ['make', 'model', 'trim', 'body', 'transmission', 'color', 'interior']

df_clean[cols] = df_clean[cols].fillna('Unknown')
```

```python
df_clean['saledate'] = pd.to_datetime(df_clean['saledate'], errors='coerce', utc=True)
```

```python
df_clean['saledate'].head()
```

```python
df_clean['saledate'].isnull().sum()
```

```python
df_clean = df_clean.dropna(subset=['saledate'])
```

```python
df_clean.to_csv("../data/processed/clean_data.csv", index=False)
```
