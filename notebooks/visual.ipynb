```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/clean_data.csv")
```

```python
df.shape
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df['sellingprice'], bins=50)
plt.title("Selling Price Distribution")
plt.savefig("../outputs/charts/price_distribution.png")
plt.show()
```

```python
top_brands = df['make'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title("Top 10 Car Brands")
plt.xticks(rotation=45)
plt.savefig("../outputs/charts/top_brands.png")
plt.show()
```

```python
brand_price = df.groupby('make')['sellingprice'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=brand_price.index, y=brand_price.values)
plt.title("Top 10 Expensive Brands")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='odometer', y='sellingprice', data=df)
plt.title("Price vs Odometer")
plt.savefig("../outputs/charts/price_vs_odometer.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='transmission', y='sellingprice', data=df)
plt.title("Transmission vs Price")
plt.savefig("../outputs/charts/transmission_price.png")
plt.show()
```

```python
df['saledate'] = pd.to_datetime(df['saledate'])

sales_trend = df.groupby(df['saledate'].dt.month)['sellingprice'].mean()

plt.figure(figsize=(10,5))
sales_trend.plot()
plt.title("Monthly Sales Trend")
plt.savefig("../outputs/charts/monthly_sales_trend.png")
plt.show()
```
